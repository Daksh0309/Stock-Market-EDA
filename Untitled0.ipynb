```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import yfinance as yf

tickers = ['TCS.NS', 'INFY.NS', 'RELIANCE.NS']

data = yf.download(tickers, start='2023-01-01', end='2024-12-31', auto_adjust=True)['Close']

print("✅ Live data loaded!")
print(data.tail())  # shows latest prices
```

```python
plt.figure(figsize=(14,6))
for col in data.columns:
    plt.plot(data.index, data[col], label=col)

plt.title('Live Stock Price Comparison')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
